```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
X1 = np.arange(5,95,5)

Y1 = np.array([279,252,250,338,308,251,216,217,214,228,237,249,264,248,250,251,253,255])
Y2 = np.array([297,269,253,253,240,244,227,289,212,219,215,220,254,233,245,236,245,231])
Y3 = np.array([216,301,192,183,377,445,636,638,364,221,219,223,224,227,226,225,231,224])


plt.plot(X1, Y1,'-o',label='SPPL')
plt.plot(X1, Y2,'-p',label='MAM')
plt.plot(X1, Y3,'-v',label='SMAM')

plt.xlabel("the number of sheep")
plt.ylabel("total steps")
plt.legend()
plt.grid()
plt.xticks(np.arange(5,95,10))
plt.show()
fig.savefig('mix.pdf',dpi=600,format='pdf')
```
